```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_gen=ImageDataGenerator(rotation_range=30,width_shift_range=0.1,height_shift_range=0.1,shear_range=0.2
                           ,zoom_range=0.2,fill_mode='nearest',horizontal_flip=True,rescale=1/255)
```

```python
dog=cv2.imread('DATA/00-puppy.jpg')
```

```python
plt.imshow(dog)
```

```python
plt.imshow(img_gen.random_transform(dog))
```
