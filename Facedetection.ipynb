```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
nadia=cv2.imread('DATA/Nadia_Murad.jpg',0)
denis=cv2.imread('DATA/Denis_Mukwege.jpg',0)
solvay=cv2.imread('DATA/solvay_conference.jpg',0)
```

```python
plt.imshow(solvay)
```

```python
faceCascade=cv2.CascadeClassifier('DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_dect(img):
    face_img=img.copy()
    face_rect=faceCascade.detectMultiScale(face_img,scaleFactor=1.15,minNeighbors=5)
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),255,thickness=5)
    return face_img
    
    
```

```python
plt.imshow(face_dect(denis))
```

```python
plt.imshow(face_dect(nadia))
```

```python
plt.imshow(face_dect(solvay))
```

```python
eyeclassifier=cv2.CascadeClassifier('DATA/haarcascades/haarcascade_eye.xml')
```

```python
def eye_detect(img):
    eye_img=img.copy()
    eye_rect=eyeclassifier.detectMultiScale(eye_img,scaleFactor=1.1,minNeighbors=5)
    for (x,y,w,h) in eye_rect:
        cv2.rectangle(eye_img,(x,y),(x+w,y+h),255,thickness=10)
    return eye_img
```

```python
plt.imshow(eye_detect(nadia))
```

```python
plt.imshow(eye_detect(denis))
```

```python
plt.imshow(eye_detect(solvay))
```

```python
capture=cv2.VideoCapture(0)

while True:
    ret,frame=capture.read(0)
    print(ret)
    if ret==True:
        frame=face_dect(frame)
        frame_eye=eye_detect(frame)
        cv2.imshow("Face detection",frame)
        k=cv2.waitKey(1)

        if k==27:
            break
        elif k==ord('e'):
            cv2.imshow("eye detect",frame_eye)
    else:
        break
        
cv2.destroyAllWindows()
    
```
